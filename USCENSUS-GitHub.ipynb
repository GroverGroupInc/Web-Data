# https://www.census.gov/data/developers/guidance/api-user-guide.html

```python
import pandas as pd
import numpy as np
import pivottablejs
import requests
import json
import csv
```

```python
response = requests.get("https://api.census.gov/data/timeseries/intltrade/exports/enduse?get=DISTRICT,DIST_NAME,ALL_VAL_MO,ALL_VAL_YR&YEAR=2016&MONTH=06")
response.status_code
```

```python
data = response.json()
```

```python
data
```

```python
infile = open('data.json', 'r')
outfile = open('data.csv', 'w')

writer = csv.writer(outfile)
for row in json.loads(infile.read()): 
    writer.writerow(row)
```

```python
with open('data.json', 'w') as outfile:
    json.dump(data, outfile)
```
